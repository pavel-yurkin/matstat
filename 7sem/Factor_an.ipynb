```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
 raw = pd.DataFrame(
[[[3.6,3.8,4.1],[2.9,3.1,3.0],[2.6,2.5,2.9]],
[[4.2,4.0,4.1],[3.3,2.9,3.2],[3.7,3.5,3.6]],
[[3.8,3.5,3.6],[3.6,3.7,3.5],[3.2,3.0,3.4]],
[[3.4,3.2,3.2],[3.4,3.6,3.5],[3.6,3.8,3.7]]]
)
```

```python
df = pd.DataFrame({'B' : ['B1', 'B2', 'B3', 'B4'],
                   'A1': [3.83, 4.1, 3.63, 3.27],
                   'A2': [3, 3.1, 3.6, 3.5],
                   'A3': [2.67, 3.6, 3.2, 3.7]})
k = 3
m = 4
n = 3
```

```python
df.loc[5] = ['SUM'] + [sum(df['A1']),sum(df['A2']),sum(df['A3'])]
df['sum'] = df.sum(axis=1)
```

```python
df.set_index('B')
```

```python
Q1 = 0
for i in range(3):
    for j in range(4):
        Q1 += df.to_numpy()[j,i+1]**2
```

```python
Q2 = sum(df.iloc[4][1:4]**2)/4
```

```python
Q3 = sum(df['sum'][:4]**2)/3
```

```python
Q4 = sum(df.iloc[4][1:4])**2/12
```

```python
Q5 = 0
for i in range(3):
    for j in range(4):
        for k in range(3):
            Q5 += raw[i][j][k]**2
```

```python
s0 = (Q1 + Q4 - Q2 - Q3)/((k-1)*(m-1))
```

```python
sA = (Q2 - Q4)/(k-1)
sA
```

```python
sB = (Q3-Q4)/(m-1)
sB
```

```python
sAB = (Q5 - n*Q1)/(m*k*(n-1))
sAB
```
