В некоторой местности в течение 300 суток регистрировалась среднесуточная температура воздуха. В итоге наблюдений было получено эмпирическое распределение, приведенное в таблице. В первом столбце указан интервал температуры в градусах, во втором столбце – частота ni , т.е. количество дней, среднесуточная температура которых принадлежит этому интервалу).

![alt text](prsn6.png "Title")

Требуется при уровне значимости α=0,05 проверить гипотезу о том, что  среднесуточная температура воздуха распределена равномерно.  

$H_0$ - Распределение среднесуточной темпиратуры воздуха равномерное.<br>
$H_1$ - Распределение среднесуточной темпиратуры воздуха НЕ равномерное.

```python
#Середина интервала
temp = [-35, -25, -15, -5, 5, 15, 25, 35]
#Частоты
freq = [25, 40, 30, 45, 40, 46, 48, 26] 
```

```python
def mean(x, freq):
    return sum([(x[i]*freq[i]) for i in range(len(x))])/sum(freq)      
```

```python
def var(x, freq, mean):
    return sum([freq[i]*(x[i] - mean)**2 for i in range(len(x))])/sum(freq)
```

```python
smpmean = mean(temp, freq)
smpvar = var(temp, freq, mean(temp, freq))
```

## Предполагаемая плотность распеделения

```python
a = smpmean - (3 * smpvar)**(0.5) 
b = smpmean + (3 * smpvar)**(0.5)
pdf = 1/(b - a)
```

## Теоретические частоты

```python
temp_f = [-40, -30, -20, -10, 0, 10, 20, 30, 40]
n_1 = sum(freq) * pdf * (temp_f[1] - a)
theory_freq = [n_1]

for i in range(len(freq) - 2):
    theory_freq.append(sum(freq) * pdf * (temp_f[i+1] -  temp_f[i]))
theory_freq.append(sum(freq) * pdf * (b - temp_f[-2]))
```

```python
theory_freq
```

```python
def pearson(f, tf):
    return sum([(f[i]-tf[i])**2/tf[i] for i in range(len(f))]) 
```

```python
#наблюдаемое значение
pearson(freq, theory_freq)
```

```python
x = [18, 10, 7, 15, 14, 11, 13]
y = [15, 20, 10, 8, 16, 10, 19] # 7, 15, 14, 29
diff = [y[i] - x[i] for i in range(len(x))]
diff
```

```python
#наблюдаемое значение
observed = lessthan0 if morethan0 > lessthan0 else morethan0 
observed
```

Уилксон: 2
![alt text](yuilk.jpg "Title")

$H_0$ - Введение новой экономической политики НЕ привело к увеличению производительности

$H_1$ - Введение новой экономической политики привело к увеличению производительности

```python
x = [52, 51, 48, 52, 54, 50, 51, 51, 52, 52, 53, 56, 51, 50]
y = [44, 47, 57, 54, 39, 65, 46, 51, 58, 46, 62, 52, 65, 47]
```

```python
yx = [[abs(y[i] - x[i]), y[i] - x[i]] for i in range(len(x))]
yx.sort() 
```

```python
dups = set()
duplicates = set()
for i in range(len(yx)):
    if yx[i][0] not in dups:
        dups.add(yx[i][0])
    else:
        duplicates.add(yx[i][0])

cnt = 2
yx[0].append(1)
for i in range(1, len(yx)):
    if yx[i][0] in duplicates and yx[i-1][0] != yx[i][0]:
        yx[i].append(cnt + 0.5)
    elif yx[i][0] in duplicates and yx[i-1][0] == yx[i][0]:
        yx[i].append(yx[i-1][2])
    elif yx[i-1][0] in duplicates and yx[i][0] not in duplicates:
        yx[i].append(cnt)
    else:
        yx[i].append(cnt)
    cnt += 1
```

```python
negative_tendention_sum = sum([yx[i][2] if yx[i][1] < 0 else 0 for i in range(len(yx))])
#наблюдаемое значение
negative_tendention_sum
```

Критическое значение при уровне значимости 0.05 и размером выборки 14 равно 25, так как наблюдаемое значение больше, чем критическое (48 > 25), гипотеза $H_0$ не отклоняется.
