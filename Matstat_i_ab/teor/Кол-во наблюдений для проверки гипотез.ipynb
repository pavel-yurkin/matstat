```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats
from tqdm import tqdm_notebook

plt.style.use('ggplot')
%matplotlib inline
```

Ошибка 1, 2 рода

Проверка гипотезы о равенстве доли 0.5 на уровне значимости a 

$H_0$ : p = 0.5 

$H_1$ : p > 0.5

```python
def prop_asy_test(x, p0=0.5, alpha=0.05):
    p = x.mean()
    z_obs = (p-p0)/np.sqrt(p0*(1-p0)/x.size)
    
    p_value = 1 - stats.norm.cdf(z_obs) #pvalue (односторонее)
    return p_value
```

```python
# Уровень значимости (ошибка первого рода) 0.05, проверка уровня значимости:

n_obs = 10**5
res = np.zeros(n_obs)

bern_rv = stats.bernoulli(p=0.5)

for i in tqdm_notebook(range(n_obs)):
    x = bern_rv.rvs(100)
    res[i] = prop_asy_test(x)
```

```python
np.mean(res < 0.05)
```

```python
# ошибка второго рода
n_obs = 10**5
res = np.zeros(n_obs)

bern_rv = stats.bernoulli(p=0.6)

for i in tqdm_notebook(range(n_obs)):
    x = bern_rv.rvs(100)
    res[i] = prop_asy_test(x)
```
