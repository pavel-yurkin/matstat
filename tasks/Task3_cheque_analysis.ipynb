```python
import numpy as np
import pandas as pd

import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
import random as rnd
plt.style.use('ggplot')  
%matplotlib inline
```

```python
df = pd.read_csv('groceries.csv', sep=',')
df.columns = ['id', 'fielddate', 'product']
print(df.shape)
df.head()
```

а) Какой товар покупался чаще всего? Сохраните название этого товара в переменную product_name.

```python
mx = df.groupby('product').count().max()[0]
product_name = df.groupby('product').agg('count')[df.groupby('product').agg('count')['fielddate'] == mx].index
product_name = str(product_name).split("'")[1]
```

б) Сколько всего уникальных заказов было сделано? Сохраните число заказов в переменную n_cnt.


```python
n_cnt = len(df.groupby('id').count().index)
```

Сделать таблицу размера "число товаров" на "число покупок", чтобы понимать какие товары покупались вместе, а какие нет. 

```python
sparse_sales = pd.pivot_table(df, 
               values='fielddate', 
               index='id', 
               columns='product', 
               fill_value=0, aggfunc='count')


sparse_sales.head()
```

в) Постройте матрицу корреляций Пирсона.

```python
sales_correlation = sparse_sales.corr(method = 'pearson')
sales_correlation
```

Какие продукты сильнее всего коррелируют с яйцами, domestic eggs (их чаще всего покупают вместе)? Сохраните название самого скоррелированного продукта в переменную top_1.

```python
top_1 = sales_correlation['domestic eggs'].sort_values().index[-2]
```

Какие продукты "мешают" купить яйца, то есть отрицательно коррелируют с их покупкой? Сохраните название продукта с самой большой отрицательной корреляцией в переменную bottom_1.

```python
bottom_1 = sales_correlation['domestic eggs'].sort_values().index[0]
```

Напишите код, который выводит самые коррелируемые товары для случайного продукта из списка unique_products.

```python
unique_products = df['product'].unique()
random_product_id = rnd.randint(1, len(unique_products))
topcorr = sales_correlation[unique_products[random_product_id]].sort_values().index[-2]
bottomcorr = sales_correlation[unique_products[random_product_id]].sort_values().index[0]

print(topcorr,'+', bottomcorr)
```

г) Какие два продукта коррелируют сильнее всего? Положите их название в лист answer


```python
max_corr = -2
answer = []
for i in sales_correlation:
    if sales_correlation[i].values[sales_correlation[i] < 1].max() > max_corr:
        max_corr = sales_correlation[i].values[sales_correlation[i] < 1].max()
for i in sales_correlation:
    if sales_correlation[i].values[sales_correlation[i] < 1].max() == max_corr:
        print(i, '+', sales_correlation[i].sort_values().index[-2])
        answer.append(i)
        answer.append(sales_correlation[i].sort_values().index[-2])
        break
```

д) Посчитайте значение метрики lift для яиц и молока ('whole milk', 'domestic eggs'). Запишите получившиеся значение метрики в переменную answer.

```python
ab = (sparse_sales[['whole milk', 'domestic eggs']] >= 1).all(axis=1).mean()
a = (sparse_sales['whole milk'] >= 1).mean()
b = (sparse_sales['domestic eggs'] >= 1).mean()
answer = ab/(a*b)
```

е) Посчитайте значение метрики для всех пар продуктов из датасета. Сохраните значения в словарик dict. В качестве ключа используете кортеж из пары продуктов.

```python
import itertools as it
comb = list(it.combinations(unique_products, 2))
food_lift = {}
for a,b in comb:
    pab = (sparse_sales[[a, b]] >= 1).all(axis=1).mean()
    pa = (sparse_sales[a] >= 1).mean()
    pb = (sparse_sales[b] >= 1).mean()
    answer = pab/((pa*pb) + 0.00000000001) #divide by zero
    food_lift[a,b] = answer
food_lift
```

Сколько пар продуктов покупали вместе хотя бы раз? Запишите ответ в переменную answer

```python
answer = 0
for v in food_lift.values():
    if v != 0:
        answer += 1
answer
```

Для какой пары продуктов метрика 𝑙𝑖𝑓𝑡 оказалась самой большой? 

```python
mx = max([i for i in food_lift.values()])
lft = 0
for k,v in food_lift.items():
    if v == mx:
        lft = k
print(lft, mx)
```

```python
print(sparse_sales[sparse_sales[lft[0]] > 0].shape[0])
print(sparse_sales[sparse_sales[lft[1]] > 0].shape[0])
```

Для какой пары продуктов метрика оказывается самой маленькой?


```python
mn = min([i for i in food_lift.values() if i != 0])
lft = 0
for k,v in food_lift.items():
    if v == mn:
        lft = k
print(lft, mn)
```
